```python
from environment.models.weather_control_fixed import WeatherControlledFixedEnv
from matplotlib import pyplot as plt
env = WeatherControlledFixedEnv(fibre_segments=3, interpolate_data=False, interpolation_values=30)

env.max_steps = 20
```

```python
while not env.done:
    env.step()

QBERS = env.get_qber()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(QBERS)
plt.legend(['QBERz', 'QBERx'])
plt.show()
```
